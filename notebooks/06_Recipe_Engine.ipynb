```python
# ==========================================================
# Import Libraries
# ==========================================================

import os
import pandas as pd
```

```python
# ==========================================================
# Project Path
# ==========================================================

PROJECT_PATH = os.path.abspath("..")

print(PROJECT_PATH)
```

```python
# ==========================================================
# Recipe Dataset Path
# ==========================================================

RECIPE_PATH = os.path.join(
    PROJECT_PATH,
    "datasets",
    "recipes",
    "recipe_data"
)

print(RECIPE_PATH)
```

```python
# ==========================================================
# Load Recipe Dataset
# ==========================================================

recipe_df = pd.read_csv(
    os.path.join(RECIPE_PATH, "RAW_recipes.csv")
)

print("Recipe Dataset Loaded Successfully!")

recipe_df.head()
```

```python
print("Recipe Dataset Shape")

recipe_df.shape
```

```python
recipe_df.columns
```

```python
# ==========================================================
# Search Banana Recipes
# ==========================================================

banana_recipes = recipe_df[

    recipe_df["name"].str.contains(
        "banana",
        case=False,
        na=False
    )

]

print("Banana Recipes Found :", len(banana_recipes))
```

```python
banana_recipes[
    [
        "name",
        "minutes",
        "n_ingredients"
    ]
].head(20)
```

```python
# ==========================================================
# Keep Recipes Between 5 and 60 Minutes
# ==========================================================

quick_recipes = banana_recipes[

    (banana_recipes["minutes"] >= 5) &
    (banana_recipes["minutes"] <= 60)

]

print("Quick Recipes Found :", len(quick_recipes))

quick_recipes[
    [
        "name",
        "minutes",
        "n_ingredients"
    ]
].head(20)
```

```python
# ==========================================================
# Keep Recipes with <= 10 Ingredients
# ==========================================================

simple_recipes = quick_recipes[
    quick_recipes["n_ingredients"] <= 10
]

print("Simple Recipes Found :", len(simple_recipes))

simple_recipes[
    [
        "name",
        "minutes",
        "n_ingredients"
    ]
].head(20)
```

```python
# ==========================================================
# Sort by Cooking Time and Ingredients
# ==========================================================

best_recipes = simple_recipes.sort_values(
    by=["minutes", "n_ingredients"],
    ascending=[True, True]
)

best_recipes[
    [
        "name",
        "minutes",
        "n_ingredients"
    ]
].head(20)
```

```python
# ==========================================================
# Top 10 Recipe Recommendations
# ==========================================================

print("=" * 60)
print("🥗 Top Recommended Banana Recipes")
print("=" * 60)

for i, (_, row) in enumerate(best_recipes.head(10).iterrows(), start=1):

    print(f"\n{i}. {row['name'].title()}")

    print(f"⏱ Cooking Time : {row['minutes']} minutes")

    print(f"🥗 Ingredients : {row['n_ingredients']}")

print("=" * 60)
```

```python
# ==========================================================
# Remove Less Healthy Recipes
# ==========================================================

UNHEALTHY_WORDS = [

    "cake",
    "cookie",
    "candy",
    "ice cream",
    "pie",
    "dessert",
    "frosting",
    "brownie",
    "chocolate",
    "cocktail"

]

healthy_recipes = best_recipes.copy()

for word in UNHEALTHY_WORDS:

    healthy_recipes = healthy_recipes[
        ~healthy_recipes["name"].str.contains(
            word,
            case=False,
            na=False
        )
    ]

print("Healthy Recipes Found :", len(healthy_recipes))

healthy_recipes[
    [
        "name",
        "minutes",
        "n_ingredients"
    ]
].head(20)
```

```python
# ==========================================================
# Top Healthy Banana Recipes
# ==========================================================

print("=" * 60)
print("🥗 Healthy Banana Recipe Recommendations")
print("=" * 60)

for i, (_, row) in enumerate(healthy_recipes.head(10).iterrows(), start=1):

    print(f"\n{i}. {row['name'].title()}")

    print(f"⏱ Cooking Time : {row['minutes']} minutes")

    print(f"🥗 Ingredients : {row['n_ingredients']}")

print("=" * 60)
```
