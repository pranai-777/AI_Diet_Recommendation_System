```python
# ==========================================================
# Import Libraries
# ==========================================================

import pandas as pd
import os
```

```python
# ==========================================================
# AI Food Classes
# ==========================================================

food_classes = [

    "bacon",
    "banana",
    "bread",
    "broccoli",
    "butter",
    "carrots",
    "cheese",
    "chicken",
    "cucumber",
    "eggs",
    "fish",
    "lettuce",
    "milk",
    "onions",
    "peppers",
    "potatoes",
    "sausages",
    "spinach",
    "tomato",
    "yogurt"

]

print(food_classes)
```

```python
# ==========================================================
# Food Mapping Table
# ==========================================================

food_mapping = {

    "bacon": "BACON",
    "banana": "BANANAS",
    "bread": "BREAD",
    "broccoli": "BROCCOLI",
    "butter": "BUTTER",
    "carrots": "CARROTS",
    "cheese": "CHEESE",
    "chicken": "CHICKEN",
    "cucumber": "CUCUMBER",
    "eggs": "EGG",
    "fish": "FISH",
    "lettuce": "LETTUCE",
    "milk": "MILK",
    "onions": "ONION",
    "peppers": "PEPPER",
    "potatoes": "POTATO",
    "sausages": "SAUSAGE",
    "spinach": "SPINACH",
    "tomato": "TOMATO",
    "yogurt": "YOGURT"

}

print(food_mapping)
```

```python
# ==========================================================
# Convert Dictionary into DataFrame
# ==========================================================

mapping_df = pd.DataFrame({

    "AI_Class": food_mapping.keys(),

    "USDA_Search": food_mapping.values()

})

mapping_df
```

```python
# ==========================================================
# Load USDA Food Database
# ==========================================================

import os
import pandas as pd

PROJECT_PATH = os.path.abspath("..")

NUTRITION_PATH = os.path.join(
    PROJECT_PATH,
    "datasets",
    "nutrition",
    "nutrition_data"
)

food_df = pd.read_csv(
    os.path.join(NUTRITION_PATH, "food.csv")
)

print("Food Database Loaded Successfully!")

food_df.head()
```

```python
# ==========================================================
# Search USDA Foods
# ==========================================================

selected_food = "banana"

search_word = food_mapping[selected_food]

results = food_df[
    food_df["description"].str.contains(
        search_word,
        case=False,
        na=False
    )
]

print("Food Selected :", selected_food)

print("USDA Search :", search_word)

print("Records Found :", len(results))
```

```python
# ==========================================================
# Show Matching Foods
# ==========================================================

results[
    [
        "fdc_id",
        "description"
    ]
].head(20)
```

```python
# ==========================================================
# Choose Best Match
# ==========================================================

best_match = results.iloc[0]

print("Best Match Found")

print()

print("Food ID :", best_match["fdc_id"])

print("Food Name :", best_match["description"])
```

```python
# ==========================================================
# Search for RAW Banana
# ==========================================================

raw_results = results[
    results["description"].str.contains(
        "RAW",
        case=False,
        na=False
    )
]

print("RAW Banana Records Found :", len(raw_results))

raw_results[
    ["fdc_id", "description"]
].head(20)
```

```python
# ==========================================================
# Choose Best Banana
# ==========================================================

if len(raw_results) > 0:

    best_match = raw_results.iloc[0]

else:

    best_match = results.iloc[0]

print("Best Food Selected")

print()

print("Food ID :", best_match["fdc_id"])

print("Food Name :", best_match["description"])
```

```python
# ==========================================================
# Food Search Function
# ==========================================================

def find_best_food(food_name):

    search_word = food_mapping[food_name]

    search_results = food_df[
        food_df["description"].str.contains(
            search_word,
            case=False,
            na=False
        )
    ]

    raw = search_results[
        search_results["description"].str.contains(
            "RAW",
            case=False,
            na=False
        )
    ]

    if len(raw) > 0:
        return raw.iloc[0]

    return search_results.iloc[0]
```

```python
banana = find_best_food("banana")

print(banana["fdc_id"])

print(banana["description"])
```
