```python
# ==========================================================
# Import Libraries
# ==========================================================

import os
import pandas as pd
```

```python
# ==========================================================
# Project Path
# ==========================================================

PROJECT_PATH = os.path.abspath("..")

print(PROJECT_PATH)
```

```python
# ==========================================================
# Nutrition Dataset Path
# ==========================================================

NUTRITION_PATH = os.path.join(
    PROJECT_PATH,
    "datasets",
    "nutrition",
    "nutrition_data"
)

print(NUTRITION_PATH)
```

```python
# ==========================================================
# Load Nutrition Tables
# ==========================================================

food_df = pd.read_csv(
    os.path.join(NUTRITION_PATH, "food.csv")
)

food_nutrient_df = pd.read_csv(
    os.path.join(NUTRITION_PATH, "food_nutrient.csv"),
    low_memory=False
)

nutrient_df = pd.read_csv(
    os.path.join(NUTRITION_PATH, "nutrient.csv")
)

print("All Nutrition Tables Loaded Successfully!")
```

```python
print("Food Table :", food_df.shape)

print("Food Nutrient Table :", food_nutrient_df.shape)

print("Nutrient Table :", nutrient_df.shape)
```

```python
# ==========================================================
# Important Nutrient IDs
# ==========================================================

IMPORTANT_NUTRIENTS = {

    1008: "Calories",
    1003: "Protein",
    1004: "Fat",
    1005: "Carbohydrates"

}

print(IMPORTANT_NUTRIENTS)
```

```python
# ==========================================================
# Banana Food ID
# ==========================================================

food_id = 790774

print(food_id)
```

```python
# ==========================================================
# Get Nutrients for Banana
# ==========================================================

banana_nutrients = food_nutrient_df[

    food_nutrient_df["fdc_id"] == food_id

]

print("Total Nutrients Found :", len(banana_nutrients))

banana_nutrients.head(15)
```

```python
# ==========================================================
# Keep Important Nutrients Only
# ==========================================================

banana_nutrients = banana_nutrients[

    banana_nutrients["nutrient_id"].isin(

        IMPORTANT_NUTRIENTS.keys()

    )

]

banana_nutrients
```

```python
# ==========================================================
# Convert Nutrient IDs to Names
# ==========================================================

banana_nutrients["Nutrient"] = banana_nutrients["nutrient_id"].map(
    IMPORTANT_NUTRIENTS
)

banana_nutrients[
    [
        "Nutrient",
        "amount"
    ]
]
```

```python
# ==========================================================
# Nutrition Dictionary
# ==========================================================

nutrition = {}

for _, row in banana_nutrients.iterrows():

    nutrition[row["Nutrient"]] = row["amount"]

nutrition
```

```python
# ==========================================================
# Display Nutrition Information
# ==========================================================

print("=" * 40)
print("🍌 Nutrition Information")
print("=" * 40)

print(f"Calories      : {nutrition['Calories']} kcal")
print(f"Protein       : {nutrition['Protein']} g")
print(f"Fat           : {nutrition['Fat']} g")
print(f"Carbohydrates : {nutrition['Carbohydrates']} g")

print("=" * 40)
```

```python
# ==========================================================
# Nutrition Function
# ==========================================================

def get_nutrition(food_id):

    # Step 1
    nutrients = food_nutrient_df[
        food_nutrient_df["fdc_id"] == food_id
    ]

    # Step 2
    nutrients = nutrients[
        nutrients["nutrient_id"].isin(
            IMPORTANT_NUTRIENTS.keys()
        )
    ]

    # Step 3
    nutrients["Nutrient"] = nutrients["nutrient_id"].map(
        IMPORTANT_NUTRIENTS
    )

    # Step 4
    nutrition = {}

    for _, row in nutrients.iterrows():

        nutrition[row["Nutrient"]] = row["amount"]

    return nutrition
```

```python
banana_nutrition = get_nutrition(790774)

banana_nutrition
```

```python
print("=" * 45)
print("🍌 Banana Nutrition Information")
print("=" * 45)

for nutrient, value in banana_nutrition.items():

    if nutrient == "Calories":

        print(f"{nutrient:<15}: {value} kcal")

    else:

        print(f"{nutrient:<15}: {value} g")

print("=" * 45)
```

```python
# ==========================================================
# Import Food Mapping Module
# ==========================================================

import sys
import os

sys.path.append(os.path.abspath("../app"))

from food_mapping import find_best_food
```

```python
# ==========================================================
# Complete Nutrition Engine
# ==========================================================

def nutrition_engine(food_name):

    # Step 1
    best_food = find_best_food(food_name, food_df)

    # Step 2
    food_id = best_food["fdc_id"]

    # Step 3
    nutrition = get_nutrition(food_id)

    # Step 4
    result = {

        "Food Name": best_food["description"],

        "Food ID": food_id,

        "Nutrition": nutrition

    }

    return result
```

```python
# ==========================================================
# Test Nutrition Engine
# ==========================================================

banana = nutrition_engine("banana")

banana
```
