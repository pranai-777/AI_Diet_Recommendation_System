```python
# ==========================================================
# Import Libraries
# ==========================================================

import os

import pandas as pd

import numpy as np
```

```python
# ==========================================================
# Project Folder
# ==========================================================

PROJECT_PATH = os.path.abspath("..")

print(PROJECT_PATH)
```

```python
# ==========================================================
# Nutrition Dataset Path
# ==========================================================

NUTRITION_PATH = os.path.join(
    PROJECT_PATH,
    "datasets",
    "nutrition",
    "nutrition_data"
)

print(NUTRITION_PATH)
```

```python
# ==========================================================
# Verify Nutrition Folder
# ==========================================================

print(os.path.exists(NUTRITION_PATH))
```

```python
# ==========================================================
# Show Dataset Files
# ==========================================================

files = os.listdir(NUTRITION_PATH)

print("Files in Nutrition Dataset:\n")

for file in files:
    print(file)
```

```python
# ==========================================================
# Read food.csv
# ==========================================================

food_df = pd.read_csv(
    os.path.join(NUTRITION_PATH, "food.csv")
)

print("Dataset Loaded Successfully!")

food_df.head()
```

```python
# ==========================================================
# Read nutrient.csv
# ==========================================================

nutrient_df = pd.read_csv(
    os.path.join(NUTRITION_PATH, "nutrient.csv")
)

print("Nutrient Dataset Loaded Successfully!")

nutrient_df.head()
```

```python
nutrient_df.shape
```

```python
nutrient_df.columns
```

```python
# ==========================================================
# Read food_nutrient.csv
# ==========================================================

food_nutrient_df = pd.read_csv(
    os.path.join(NUTRITION_PATH, "food_nutrient.csv")
)

print("Food Nutrient Dataset Loaded Successfully!")

food_nutrient_df.head()
```

```python
food_nutrient_df.shape
```

```python
food_nutrient_df.columns
```

```python
# ==========================================================
# Important Nutrients
# ==========================================================

important_nutrients = [
    "Protein",
    "Total lipid (fat)",
    "Carbohydrate, by difference",
    "Energy"
]

filtered_nutrients = nutrient_df[
    nutrient_df["name"].isin(important_nutrients)
]

filtered_nutrients
```

```python
# ==========================================================
# Search for Banana
# ==========================================================

banana = food_df[
    food_df["description"].str.contains(
        "banana",
        case=False,
        na=False
    )
]

print("Total Banana Records Found:", len(banana))

banana.head(10)
```
