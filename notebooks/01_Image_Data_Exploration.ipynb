```python
# ==========================================================
# Import Required Libraries
# ==========================================================

import os
import random

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

from PIL import Image
```

```python
# ==========================================================
# Find Project Folder
# ==========================================================

PROJECT_PATH = os.path.abspath("..")

print("Project Folder:")
print(PROJECT_PATH)
```

```python
# ==========================================================
# Dataset Paths
# ==========================================================

IMAGE_DATASET = os.path.join(
    PROJECT_PATH,
    "datasets",
    "food_images",
    "food_images"
)

NUTRITION_DATASET = os.path.join(
    PROJECT_PATH,
    "datasets",
    "nutrition",
    "nutrition_data"
)

RECIPE_DATASET = os.path.join(
    PROJECT_PATH,
    "datasets",
    "recipes",
    "recipe_data"
)

print("Image Dataset:")
print(IMAGE_DATASET)

print("\nNutrition Dataset:")
print(NUTRITION_DATASET)

print("\nRecipe Dataset:")
print(RECIPE_DATASET)
```

```python
# ==========================================================
# Verify Dataset Paths
# ==========================================================

print("Image Dataset Found :", os.path.exists(IMAGE_DATASET))
print("Nutrition Dataset Found :", os.path.exists(NUTRITION_DATASET))
print("Recipe Dataset Found :", os.path.exists(RECIPE_DATASET))
```

```python
# ==========================================================
# Explore Image Dataset
# ==========================================================

print("Folders inside the Image Dataset:")
print(os.listdir(IMAGE_DATASET))
```

```python
# ==========================================================
# Save Dataset Folder Names
# ==========================================================

train_path = os.path.join(IMAGE_DATASET, "train")
test_path = os.path.join(IMAGE_DATASET, "test")
val_path = os.path.join(IMAGE_DATASET, "val")

print("Training Folder:")
print(train_path)

print("\nTesting Folder:")
print(test_path)

print("\nValidation Folder:")
print(val_path)
```

```python
# ==========================================================
# Verify Train, Test and Validation Folders
# ==========================================================

print("Train Folder Exists :", os.path.exists(train_path))
print("Test Folder Exists  :", os.path.exists(test_path))
print("Validation Exists   :", os.path.exists(val_path))
```

```python
# ==========================================================
# Food Categories (Classes)
# ==========================================================

food_classes = sorted(os.listdir(train_path))

print("Food Categories:\n")

for food in food_classes:
    print(food)
```

```python
# ==========================================================
# Number of Food Categories
# ==========================================================

print("Total Food Categories:", len(food_classes))
```

```python
# ==========================================================
# Display Food Categories
# ==========================================================

print("Food Categories\n")

for index, food in enumerate(food_classes, start=1):
    print(f"{index}. {food}")
```

```python
# ==========================================================
# Count Images in Each Food Category
# ==========================================================

print("Number of Images in Each Food Category\n")

for food in food_classes:
    
    folder = os.path.join(train_path, food)
    
    total_images = len(os.listdir(folder))
    
    print(f"{food:<12} : {total_images}")
```

```python
# ==========================================================
# Display One Sample Image
# ==========================================================

# Choose the first food category
food = food_classes[0]

# Create the folder path
food_folder = os.path.join(train_path, food)

# Get the first image name
image_name = os.listdir(food_folder)[0]

# Create full image path
image_path = os.path.join(food_folder, image_name)

# Open image
image = Image.open(image_path)

# Display image
plt.figure(figsize=(5,5))
plt.imshow(image)
plt.title(food)
plt.axis("off")

plt.show()
```

```python
# ==========================================================
# Display Random Food Images
# ==========================================================

plt.figure(figsize=(12,12))

for i in range(9):

    # Pick a random food class
    food = random.choice(food_classes)

    # Open that folder
    folder = os.path.join(train_path, food)

    # Pick a random image
    image_name = random.choice(os.listdir(folder))

    # Open image
    image = Image.open(os.path.join(folder, image_name))

    # Display
    plt.subplot(3,3,i+1)

    plt.imshow(image)

    plt.title(food)

    plt.axis("off")

plt.tight_layout()

plt.show()
```

```python
# ==========================================================
# Check Image Size
# ==========================================================

print("Image Size:", image.size)
```

```python
# ==========================================================
# Check for Corrupted Images
# ==========================================================

broken_images = []

for food in food_classes:

    folder = os.path.join(train_path, food)

    for image_name in os.listdir(folder):

        image_path = os.path.join(folder, image_name)

        try:

            img = Image.open(image_path)

            img.verify()

        except:

            broken_images.append(image_path)

print("Total Broken Images:", len(broken_images))
```

```python
# ==========================================================
# Check Image Dimensions
# ==========================================================

image_sizes = []

for food in food_classes:

    folder = os.path.join(train_path, food)

    for image_name in os.listdir(folder):

        image_path = os.path.join(folder, image_name)

        with Image.open(image_path) as img:
            image_sizes.append(img.size)

print("Total Images Checked:", len(image_sizes))

unique_sizes = set(image_sizes)

print("Number of Different Image Sizes:", len(unique_sizes))

print("\nFirst 10 Different Sizes:")

for size in list(unique_sizes)[:10]:
    print(size)
```

```python
# ==========================================================
# Store Number of Images in Each Food Category
# ==========================================================

image_count = {}

for food in food_classes:

    folder = os.path.join(train_path, food)

    image_count[food] = len(os.listdir(folder))

print(image_count)
```

```python
# ==========================================================
# Bar Chart of Food Categories
# ==========================================================

plt.figure(figsize=(12,6))

plt.bar(
    image_count.keys(),
    image_count.values()
)

plt.title("Number of Training Images per Food Category")

plt.xlabel("Food Category")

plt.ylabel("Number of Images")

plt.xticks(rotation=45)

plt.tight_layout()

plt.show()
```

```python
# ==========================================================
# Pie Chart
# ==========================================================

plt.figure(figsize=(10,10))

plt.pie(
    image_count.values(),
    labels=image_count.keys(),
    autopct="%1.1f%%",
    startangle=90
)

plt.title("Distribution of Food Categories")

plt.show()
```
